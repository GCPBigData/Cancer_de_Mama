```python
# importando biblioteca
#import pandas as pd
import numpy as np
import joblib as jb
import warnings
warnings.filterwarnings("ignore")
```

# EXECUTANDO MODELO

```python
## LER OBJETO QUE CONTEM O MODELO
arquivo_modelo = jb.load('classificacao_cancer_mama.joblib')
arquivo_modelo
```

```python
## RETORNANDO O MODELO PARA UM OBJETO
modelo = arquivo_modelo['model']
modelo
```

```python
print(dir(modelo))
```

```python
# CRIANDO NOVOS DADOS
texture_mean = 24.54
smoothness_mean = 0.05263	
compactness_mean = 0.04362
symmetry_mean = 0.1587
fractal_dimension_mean = 0.05884

novo_dados = np.array([[texture_mean ,smoothness_mean,compactness_mean,symmetry_mean,fractal_dimension_mean]])
novo_dados
```

```python
## FAZENDO PREDIÇÃO NOS NOVOS DADOS
resultado = modelo.predict(novo_dados)
if resultado[0] == 0:
  print ("Com o dados atuais do paciente, o possível tipo de câncer é Maligno")
else:
  print ("Com o dados atuais do paciente, o possível tipo de câncer é Beligno")
```
